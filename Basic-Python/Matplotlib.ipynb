### Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))
# plot to each different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(n_b_p.keys(),n_b_p.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig ,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(n_b_p.keys(),n_b_p.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use pyplot vs oo method
* when we are ploting quick use pyplot
* but when we want advanced plot we use oo method

```python
over_fifty = heart_disease[heart_disease["age"]>50]
```

```python
over_fifty
```

```python
over_fifty.plot.scatter(x="age",
                    y="chol",
                    c="target");
```

```python
## OO method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_fifty.plot.scatter(x="age",
                       y="chol",
                       c="target",
                       ax=ax);
ax.set_xlim([45,100]); ## it limits the axes limits
```

```python
# OO method form scratch
# Now the c= which means color will change the color of the plot depending on the target col of the df
fig,ax= plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_fifty["age"],
                    y=over_fifty["chol"],
                    c=over_fifty["target"]);
ax.set(title="Heart Disease and Cholestrol",
      xlabel="age",
      ylabel="cholesptrol");
ax.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax.axhline(over_fifty["chol"].mean() , linestyle="--");
```

```python
over_fifty.head()
```

```python
# Subplot of chol age, thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter = ax0.scatter(x=over_fifty["age"],
                    y=over_fifty["chol"],
                    c=over_fifty["target"]);
ax0.set(title="Heart Disease and Cholestrol",
        xlabel="Age",
      ylabel="cholestrol");
ax0.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax0.axhline(over_fifty["chol"].mean() , linestyle="--");

scatter = ax1.scatter(x=over_fifty["age"],
                    y=over_fifty["thalach"],
                    c=over_fifty["target"]);
ax1.set(title="Heart Disease and Max Heart Rate",
      ylabel="thalach");
ax1.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax1.axhline(over_fifty["thalach"].mean() , linestyle="--");
```

```python
# Create some dummy data
x = np.random.randn(10,4)
df= pd.DataFrame(x)
df.columns=["a","b","c","d"]
df
```

```python
ax =df.plot.bar()
```

```python
# Customize our plot using the set method
ax = df.plot(kind="bar")
# add some lables and titles
ax.set(title="Random number bar graph from the data frame",
      xlabel="Row Number",
      ylabel="Random number")
# Make the legend visible
ax.legend().set_visible(True) #  but the legend is already visible
```

```python
# set the style
plt.style.use('seaborn-whitegrid')
# OO method form scratch
# Now the c= which means color will change the color of the plot depending on the target col of the df
fig,ax= plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_fifty["age"],
                    y=over_fifty["chol"],
                    c=over_fifty["target"],
                    cmap="winter");
ax.set(title="Heart Disease and Cholestrol",
      xlabel="age",
      ylabel="cholesptrol");
ax.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax.axhline(over_fifty["chol"].mean() , linestyle="--");
```

```python
# Subplot of chol age, thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter = ax0.scatter(x=over_fifty["age"],
                    y=over_fifty["chol"],
                    c=over_fifty["target"],
                     cmap="winter");
ax0.set(title="Heart Disease and Cholestrol",
      ylabel="cholestrol");
# change the axo axes limits
ax0.set_xlim(50,80)
ax0.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax0.axhline(over_fifty["chol"].mean() , linestyle="--");

scatter = ax1.scatter(x=over_fifty["age"],
                    y=over_fifty["thalach"],
                    c=over_fifty["target"],
                     cmap="winter");
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
      ylabel="thalach");
ax1.set_xlim=(50,80)
ax1.set_ylim(60,200)
ax1.legend(*scatter.legend_elements(), title="Target"); # it will go to the scatter created above than find c from where it takes the leagend values which will led it to the target column

ax1.axhline(over_fifty["thalach"].mean() , linestyle="--");
```
